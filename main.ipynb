```python
# Keep numeric types intact
df['director'].fillna("Unknown", inplace=True)
df['cast'].fillna("Unknown", inplace=True)
df['country'].fillna("Unknown", inplace=True)
df['rating'].fillna("Not Rated", inplace=True)
df['duration'].fillna("Unknown", inplace=True)
```

```python
# Convert the 'date_added' column to datetime format. 
# If a value can't be parsed, it will be set to NaT (missing date).
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df['date_added']
```

```python
df = df.rename(columns={
    'show_id': 'Film_ID',
    'type': 'Type_Of_Entertainment',
    'title': 'Title',
    'director': 'Director',
    'cast': 'Casts',
    'country': 'Country',
    'date_added': 'Added_on_site',
    'release_year': 'Film_Releasing_Year',
    'rating': 'Film_Rating',
    'duration': 'Duration_Of_Film',
    'listed_in': 'Genres',
    'description': 'Description'
})[
    ['Film_ID','Type_Of_Entertainment','Title','Director','Casts','Country',
     'Added_on_site','Film_Releasing_Year','Film_Rating',
     'Duration_Of_Film','Genres','Description']
]
```

```python
df
```

```python
# (a) Entertainment type as category
df['Type_Of_Entertainment'] = df['Type_Of_Entertainment'].astype('category')

# (b) Release year as int
df['Film_Releasing_Year'] = df['Film_Releasing_Year'].astype(int)

# (c) Extract year and month from Added_on_site
df['Year_Added'] = df['Added_on_site'].dt.year
df['Month_Added'] = df['Added_on_site'].dt.month

# (d) Handle Duration: extract numbers and TV/Movie flag
df['Duration_Number'] = df['Duration_Of_Film'].str.extract('(\d+)')
df['Duration_Number'] = pd.to_numeric(df['Duration_Number'], errors='coerce')

df['Is_TV_Show'] = df['Duration_Of_Film'].str.contains("Season").astype(int)

# (e) Genres: split into list
df['Genres'] = df['Genres'].str.split(", ")
```

```python
df = df.sort_values(by='Film_Rating', ascending=False)
```

```python
df
```

```python
df.index = range(1,len(df)+1)
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
df.index = df.index + 1
df.insert(0, 'Index', df.index)
```
